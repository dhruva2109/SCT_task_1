```
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (update the file path as needed)
file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_3401680.csv'
data = pd.read_csv(file_path, skiprows=4)  # Skip the first 4 rows as they usually contain metadata

# Extract data for Afghanistan
afghanistan_data = data[data['Country Name'] == 'Afghanistan'].iloc[:, 4:-1].T  # Transpose to make years as rows
afghanistan_data.columns = ['Population']  # Rename the column
afghanistan_data.index = afghanistan_data.index.astype(int)  # Convert index (years) to integer

# Plot the population trend for Afghanistan
plt.figure(figsize=(10, 6))
plt.plot(afghanistan_data.index, afghanistan_data['Population'], marker='o', color='b', label='Afghanistan')
plt.title('Population Trend for Afghanistan (1960-2023)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
